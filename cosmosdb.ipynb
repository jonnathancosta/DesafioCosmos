```python
# 3 ATUALIZACAO DE ESTOQUE APÓS UM PEDIDO 
#Quando um pedido é realizado, o estoque do produto precisa ser atualizado. Escreva uma consulta para reduzir o estoque em 1 unidade para o 
# produto_id especificado em um pedido.

client = MongoClient(connection_string)

# Nome do banco de dados e coleções
db = client['jonnathan_costa_DB']
produtos = db['produtos']
pedidos = db['pedidos']

# Dados do pedido (exemplo)
pedido = {
    "pedido_id": "ped011",
    "usuario_id": "user002",
    "produto_id": "prod004",  # produto a ser pedido
    "quantidade": 2  # quantidade do produto
}

# 1. Inserir o pedido na coleção de pedidos
pedidos.insert_one(pedido)
print(f"Pedido {pedido['pedido_id']} inserido.")

# 2. Atualizar o estoque do produto, diminuindo pela quantidade pedida
produto_id = pedido["produto_id"]
quantidade_pedida = pedido["quantidade"]

# Buscar o produto para verificar o estoque atual
produto = produtos.find_one({"produto_id": produto_id})

if produto:
    estoque_atual = produto["estoque"]
    if estoque_atual >= quantidade_pedida:
        # Atualizar o estoque, diminuindo pela quantidade pedida
        produtos.update_one(
            {"produto_id": produto_id},  # Condição para encontrar o produto
            {"$inc": {"estoque": -quantidade_pedida}}  # Reduzir o estoque
        )
        print(f"Estoque do produto {produto_id} atualizado. Novo estoque: {estoque_atual - quantidade_pedida}")
    else:
        print(f"Estoque insuficiente para o pedido. Estoque atual: {estoque_atual}")
else:
    print(f"Produto com produto_id {produto_id} não encontrado.")

# Fechar a conexão
client.close()
```

```python
#4 ANALISE DE VENDAS POR CATEGORIA
#Usando operações de agregação, calcule o número total de pedidos para cada categoria de produto.
# Exiba a categoria e a quantidade_total de pedidos.


client = MongoClient(connection_string)

# Nome do banco de dados e coleções
db = client['jonnathan_costa_DB']
produtos = db['produtos']
pedidos = db['pedidos']

# Pipeline de agregação
pipeline = [
    # 1. Realizar a junção com a coleção de produtos para pegar a categoria do produto
    {
        "$lookup": {
            "from": "produtos",               # Nome da coleção de produtos
            "localField": "produto_id",        # Campo de chave na coleção de pedidos
            "foreignField": "produto_id",      # Campo de chave na coleção de produtos
            "as": "produto_info"               # Nome do campo onde os dados de produtos serão armazenados
        }
    },
    # 2. Desconstruir o array "produto_info" para poder acessar os campos da categoria
    {
        "$unwind": "$produto_info"
    },
    # 3. Agrupar por categoria e contar o número total de pedidos em cada categoria
    {
        "$group": {
            "_id": "$produto_info.categoria",   # Agrupar pela categoria do produto
            "quantidade_total_pedidos": {"$sum": 1}  # Contar o número de pedidos em cada categoria
        }
    },
    # 4. Ordenar as categorias em ordem decrescente de número de pedidos
    {
        "$sort": {"quantidade_total_pedidos": -1}  # Ordenar de forma decrescente
    }
]

# Executar o pipeline de agregação
resultado = pedidos.aggregate(pipeline)

# Exibir os resultados
print("\nAnálise de Vendas por Categoria:")
for categoria in resultado:
    print(f"Categoria: {categoria['_id']}, Quantidade Total de Pedidos: {categoria['quantidade_total_pedidos']}")

# Fechar a conexão
client.close()
```

```python
# 5 Top 3 Produtos Mais Vendidos:

#Identifique os três produtos mais vendidos e liste o nome e o total_vendido de cada um. 
# Use agregações para somar as quantidades dos produtos nos pedidos.

client = MongoClient(connection_string)

# Nome do banco de dados e coleções
db = client['jonnathan_costa_DB']
pedidos = db['pedidos']
produtos = db['produtos']

# Pipeline de agregação
pipeline = [
    # 1. Agrupar os pedidos por produto_id e somar as quantidades
    {
        "$group": {
            "_id": "$produto_id",  # Agrupar pelo produto_id
            "quantidade_total_vendida": {"$sum": "$quantidade"}  # Somar a quantidade vendida de cada produto
        }
    },
    # 2. Ordenar os produtos pela quantidade total vendida, de forma decrescente
    {
        "$sort": {"quantidade_total_vendida": -1}
    },
    # 3. Limitar os resultados para os 3 produtos mais vendidos
    {
        "$limit": 3
    },
    # 4. Fazer um lookup para obter o nome do produto a partir da coleção 'produtos'
    {
        "$lookup": {
            "from": "produtos",  # Nome da coleção de produtos
            "localField": "_id",  # Campo de chave na coleção de pedidos (produto_id)
            "foreignField": "produto_id",  # Campo de chave na coleção de produtos
            "as": "produto_info"  # Nome do campo que conterá os dados do produto
        }
    },
    # 5. Desconstruir o array 'produto_info' (pois será um array com um único item)
    {
        "$unwind": "$produto_info"
    },
    # 6. Projetar a saída final com o nome do produto e a quantidade total vendida
    {
        "$project": {
            "nome": "$produto_info.nome",  # Nome do produto
            "quantidade_total_vendida": 1,  # Quantidade total vendida
            "_id": 0  # Não incluir o campo _id
        }
    }
]

# Executar a agregação
resultado = pedidos.aggregate(pipeline)

# Exibir os resultados
print("\nTop 3 Produtos Mais Vendidos:")
for produto in resultado:
    print(f"Produto: {produto['nome']}, Quantidade Vendida: {produto['quantidade_total_vendida']}")

# Fechar a conexão
client.close()


```
